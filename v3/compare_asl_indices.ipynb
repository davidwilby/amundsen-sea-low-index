# Comparing ASL indices v2 and v3 using ERA-Interim

* v2: written in IDL (Hosking et al., 2016)
* v3: writtin in python, using scikit-image ([see here](https://github.com/scott-hosking/amundsen_sea_low_index))

```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
import xarray as xr
```

### Read in ASL index (v2)
* process date to 'YYYY-MM-01'
* rename columns so we can merge indices

```python
asli_v2 = pd.read_csv(
            'https://bas.ac.uk/data/absl/ASL-index-Version2-Monthly-ERA-Interim_Hosking2016.txt', 
            skiprows=21, delim_whitespace=True)

time = [ datetime.date(int(row.Year), int(row.Month), 1).strftime('%Y-%m-%d') for i, row in asli_v2.iterrows() ]

asli_v2['time'] = time

asli_v2['RelCenPres'] = asli_v2['ActCenPres'] - asli_v2['SectorPres']

for col in ['lat','lon','ActCenPres','SectorPres','RelCenPres']:
    asli_v2.rename({col: col+'2'}, axis=1, inplace=True)

asli_v2 = asli_v2.drop(labels=['Year', 'Month'], axis=1)

asli_v2.iloc[-3:]
```

### Read in ASL index (v3)
* rename columns so we can merge indices

```python
asli_v3 = pd.read_csv('era-interim/asli_v3.20200107-era-interim.csv', 
                         comment='#')

for col in ['lat','lon','ActCenPres','SectorPres','RelCenPres']:
    asli_v3.rename({col: col+'3'}, axis=1, inplace=True)
    
asli_v3.iloc[-3:]
```

### Merge two dataframes in time so that we can compare them

```python
df = pd.merge(asli_v3,asli_v2, on='time')
```

```python
df.iloc[-3:]
```

### Focus in on months where we see large differences in ASL lon and lat

```python
df['lon_diff'] = np.abs(df['lon3'] - df['lon2'])
df['lat_diff'] = np.abs(df['lat3'] - df['lat2'])
```

```python
df_lon_sort = df.sort_values('lon_diff', axis=0, ascending=False).reset_index(drop=True)
df_lon_sort.iloc[0:6]
```

```python
df_lat_sort = df.sort_values('lat_diff', axis=0, ascending=False).reset_index(drop=True)
df_lat_sort.iloc[0:6]
```

```python
### read in ERA-Interim gridded data to plot maps
da = xr.open_dataset('../INDATA/ERAI/erai_sfc_monthly.nc').msl / 100.

mask = xr.open_dataset('../INDATA/ERAI/erai_invariant.nc').lsm.squeeze()
da_mask = da.where(mask == 0)

def slice_region(da, region, boarder=8):
    da = da.sel( latitude=slice(region['north']+boarder,region['south']-boarder), 
                longitude=slice(region['west']-boarder,region['east']+boarder))
    return da

# region of interest (asl sector)
asl_region = {'west':170., 'east':298., 'south':-80., 'north':-60.}

da = slice_region(da, asl_region)
da_mask = slice_region(da_mask, asl_region)
```

```python
def draw_regional_box( region, transform=None ):
    '''
    Draw box around a region on a map
    region is a dictionary with west,east,south,north
    '''

    if transform == None:
        transform = ccrs.PlateCarree()

    plt.plot([region['west'], region['west']], [region['south'],region['north']], 
                 'k-', transform=transform, linewidth=1)
    plt.plot([region['east'], region['east']], [region['south'],region['north']], 
                 'k-', transform=transform, linewidth=1)
    
    for i in range( np.int(region['west']),np.int(region['east']) ): 
        plt.plot([i,i+1], [region['south'],region['south']], 'k-', transform=transform, linewidth=1)
        plt.plot([i,i+1], [region['north'],region['north']], 'k-', transform=transform, linewidth=1)
```

```python
### plots sorted by difference in longitude

plt.figure(figsize=(15,10))

for i in range(0,6):

    df1 = df_lon_sort.iloc[i]
    da1 = da_mask.sel(time=df1.time+'T00:00:00.000000000')
    
    ax = plt.subplot( 2,3,i+1, 
                        projection=ccrs.Stereographic(central_longitude=0., 
                                                      central_latitude=-90.) )

    ax.set_extent([160,305,-85,-55], ccrs.PlateCarree())

    result = da1.plot.pcolormesh('longitude', 'latitude', cmap='Reds', 
                                        transform=ccrs.PlateCarree(), 
                                        add_colorbar=False, 
                                        levels=np.linspace(np.nanmin(da1.values), np.nanmax(da1.values), 20) )

    draw_regional_box(asl_region)

    plt.plot(df1['lon2'], df1['lat2'], 'kx', transform=ccrs.PlateCarree() )
    plt.plot(df1['lon3'], df1['lat3'], 'k+', transform=ccrs.PlateCarree() )
```

```python
# plots sorted by difference in latitude

plt.figure(figsize=(15,10))

for i in range(0,6):

    df1 = df_lat_sort.iloc[i]
    da1 = da_mask.sel(time=df1.time+'T00:00:00.000000000')
    
    ax = plt.subplot( 2,3,i+1, 
                        projection=ccrs.Stereographic(central_longitude=0., 
                                                      central_latitude=-90.) )

    ax.set_extent([165,305,-85,-55], ccrs.PlateCarree())

    result = da1.plot.pcolormesh('longitude', 'latitude', cmap='Reds', 
                                        transform=ccrs.PlateCarree(), 
                                        add_colorbar=False, 
                                        levels=np.linspace(np.nanmin(da1.values), np.nanmax(da1.values), 20) )

    asl_region = {'west':170., 'east':298., 'south':-80., 'north':-60.}
    draw_regional_box(asl_region)

    plt.plot(df1['lon2'], df1['lat2'], 'kx', transform=ccrs.PlateCarree() )
    plt.plot(df1['lon3'], df1['lat3'], 'k+', transform=ccrs.PlateCarree() )
```
