```python
import xarray as xr
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import numpy as np
```

```python
lsm = xr.open_dataset('data/era5_lsm.nc')['lsm'].squeeze()
# lsm = xmslr.open_dataset('data/invariant.nc')['lsm'].squeeze()
```

```python
msl = xr.open_dataarray("data/ERA5/monthly/era5_mean_sea_level_pressure_monthly_1979.nc")[0]
```

```python
fig = plt.figure(figsize=[12,5])

ax = fig.add_subplot(111, projection=ccrs.Stereographic(central_longitude=0., central_latitude=-90.))

msl.where(lsm<0.5).plot.contourf(ax=ax,
                   levels=np.arange(98000, 104000, 100),
                   extend='max',
                   transform=ccrs.PlateCarree())
ax.set_extent([-180,180,-90,-50], ccrs.PlateCarree())
ax.coastlines()

plt.show()
```
